# **RNN (recurrent Neural Network)**

A Recurrent Neural Network (RNN) is a type of neural network designed to work with sequential data.

Unlike normal feedforward neural networks, RNNs have memory: they use previous outputs (hidden states) as inputs for the next step.

# **Why RNN?**

- Normal Neural Networks assume inputs are independent.

 - But in sequences, current data depends on previous context (e.g., in a sentence, the meaning of a word depends on the words before it).

- RNNs keep track of past information using hidden states.

Example

Lets say we want to predict the next word in a sentence:

Input: "I love deep"

RNN processes each word in order: "I"  "love"  "deep"

The hidden state keeps track of context.

Finally, it predicts the next word: "learning".

Types of RNNs

Vanilla RNN  basic type, suffers from vanishing gradient problem.

LSTM (Long Short-Term Memory)  special RNN with gates to remember long dependencies.

GRU (Gated Recurrent Unit)  simpler than LSTM, but powerful.

Applications of RNN

Natural Language Processing (NLP): text generation, translation, sentiment analysis.

Speech Recognition

Time-Series Forecasting (stock prices, weather prediction).
